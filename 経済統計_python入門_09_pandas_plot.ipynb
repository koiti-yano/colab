```python
# 最初にちょっと重要な呪文を唱えます。
# 最初は何をしているのかわからなくても構いません。
import pandas as pd
# matplotlibをimpoertする必要はありません。
```

グラフを作るため、データ（CSVファイル）をインターネットから読み込みます。ここではread_csvというメソッドを使ってインターネット上にあるデータをPythonに読み込み、ice_dataという名前で保存します。head()というメソッドはデータを上から5行表示するメソッドです。

```python
ice_data = pd.read_csv("https://www.dropbox.com/s/2ky13qi387j8b05/kion_iceCream.csv?raw=1",encoding="utf_8")
ice_data.head()
```

Colabは日本語に対応していないため、列名を英語に書き換えます。inplace=Trueとすれば、オブジェクト本体の列名が書き換えられます。

参考<br>
https://note.nkmk.me/python-pandas-dataframe-rename/


```python
ice_data.rename(columns={'年月': 'Date', '気温': 'Temperature', 'アイスクリーム支出金額': 'Icecream'}, inplace=True)
ice_data.head()
```

次にデータフレームice_dataのIcecream（アイスクリーム支出金額）をプロットしてみましょう。Pandasの折れ線グラフでは縦軸になる列名をy=「列名」と指定します。

Icecream（アイスクリーム支出金額）を折れ線グラフにする場合、y="Icecream"と指定し、以下のように書きます。

```python
ice_data.plot(y="Icecream")
```

次に気温とアイスクリーム支出金額の２つのデータを使って散布図を作ってみましょう。散布図ではplot.scatter()を使い、横軸になる列名をx=「列名」、縦軸になる列名をy=「列名」を指定します。

```python
ice_data.plot.scatter(x='Temperature', y='Icecream')
```
