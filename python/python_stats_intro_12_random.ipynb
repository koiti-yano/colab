**離散一様乱数**

離散一様乱数を発生させるメソッドはrandom_integers()です。以下のようにパラメーターを指定します。

random_integers(low=下限, high=上限, size=発生させる乱数の個数)

たとえば、0から9までの整数から離散一様乱数を100個発生させる場合、以下のように入力します。

```python
numb_int = np.random.random_integers(low=0, high=9, size=100)
print(numb_int)
```

ヒストグラムを描いてみましょう。

```python
plt.hist(numb_int)
```

**連続一様乱数**

連続一様乱数を発生させるメソッドはuniform()です。以下のようにパラメーターを指定します。

unifrom(low=下限, high=上限, size=発生させる乱数の個数)

たとえば、下限が0、上限が1である連続一様乱数を10000個発生させる場合、以下のように入力します。

```python
numb_unif = np.random.uniform(low=0,high=1, size=10000)
plt.hist(numb_unif)
```

**正規乱数**

正規乱数を発生させるメソッドはnormal()です。

$X \sim \mathcal{N}(\mu,\,\sigma^{2})$

以下のようにパラメーターを指定します。

normal(loc=平均, scale=標準偏差, size=発生させる乱数の個数)

たとえば、平均が0、標準偏差が1である正規乱数を10000個発生させる場合、以下のように入力します。

```python
numb_normal = np.random.normal(loc=0, scale=1, size=10000)
plt.hist(numb_normal)
```

**標準正規乱数**

標準正規乱数を発生させる場合は、standard_normal()を使います。

standard_normal(size=発生させる乱数の個数)

たとえば、標準正規乱数を10000個発生させる場合、以下のように入力します。

> 標準正規乱数を発生させる場合、standard_normal(loc=0, scale=1)・normal()どちらを使っても構いません（正しく設定されいれば）。



```python
numb_st_normal = np.random.standard_normal(size=10000)
plt.hist(numb_st_normal)
```

補足：標準正規乱数を発生させるメソッドとしてrandn()も使えます。使っていけないわけではありませんが、個人的にはstandard_normal()を使うほうがいいのではないかと思っています（randnはMatlabというプログラム言語のrandn関数との互換性のためにあるだけなので）。
