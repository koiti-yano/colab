**「等しい」を表すには？**<br>
たとえば、変数aに5を代入して、その後、「あれ？変数aに5がちゃんと入っているかな？」と疑問に思ったとしましょう。どのように確認したらいいでしょうか？
<br>

まず、変数aに5を代入してみましょう。=を使います。

```python
a = 5
```

これで変数aに数字の5が代入されました。続いて、「変数aに数字の5が代入されているか」を確認してみましょう。

この場合==（等号を2つ）を使います。==は「（左右の）両辺が等しい」を意味します。ちょっと使ってみましょう。

```python
a == 5
```

True（真）と表示されました。これは「（変数aに5が代入されているので）a==5は正しい」ことを意味します。

> **比較演算子** ==, <, > <=, >=などを比較演算子といいます（最初は難しい名前を覚える必要はありません！）

*   \<：小（左辺が右辺より小）
*   \>：大（左辺が右辺より大）
*   \<\=：以上（左辺が右辺以下。注：左辺＝右辺を含む）
*   \>=：以下（左辺が右辺以上。注：左辺＝右辺を含む）
*   \!\=：等しくない（両辺が等しくない）

ちょっと分かりにくい「\<\=」と「\!\=」を使ってみましょう。





```python
a = 5
b = 6
c = 7
a <= 6
```

```python
b <= 6
```

```python
c <= 6
```

変数cには7が入っていますから「c <= 6」のときにはFalse（偽）と出てきます。

また、「等しくない」ことを表す「=!」もやってみましょう。

```python
c != 6
```

Trueと出ました。ちょっとわかりにくいですね。意味としては以下のようになります。

1.   変数cには7が入っている
2.   cに入っている7と数字の6は等しくない
3.   「c!=6」は「等しくない」という関係を正しく表している
4.   だから「c!=6」という式はTrue（真）［等しくないという関係を正しく表している］

ということですね。

**朗報：言葉にも==を使える**

さて、ここまでは数字に==（等しい）が使えることを説明してきました。実は言葉（正確には文字列といいます）にも==（等しい）が使えます。

「何を言っているんだ？」と思う人には実例を見たほうが早いでしょう。

```python
joken = "晴れ" # jokenに「晴れ」という文字列（言葉）を代入しています
```

jokenに「晴れ」という文字列（言葉）が代入されているとき、joken=="晴れ"を実行するとどうなるでしょうか？「わからない」？やってみましょう。

```python
joken == "晴れ"
```

jokenには「晴れ」が入っていますから、「jokenに入っている『晴れ』は文字列『晴れ』と等しい」は真（正しい）ですね。

次にjoken=="雨"を実行するとどうなるでしょうか？こちらも早速やってみましょう。

```python
joken == "雨"
```

jokenには「晴れ」が入っているので、「雨とは等しくない」ですね。なので、偽（正しくない）が出力されます。

このように文字列（言葉）についても==が使えるわけです。

「そんなの何の役に立つのか？」と思うかもしれませんが、次の例で早速役に立ちます。

**条件分岐とif文**

「もし、明日、晴れなら傘を持たずに外出する、雨なら傘を持って外出する」。

*   条件：明日が晴れ→結果：傘を持たずに外出
*   条件：明日が雨→結果：傘を持って外出

「雨なら傘を持って外出する」なんて、ごく当たり前の話ですが、このように条件によってやることが異なる場合をプログラミングでは条件分岐といいます。

「もし◯◯なら」を英語ではifを使うので、プログラミングでもifを使います。

**勇者の選択を考える**

ダンジョンでドラゴンに遭遇した勇者の選択を考えましょう。

1.   戦う
2.   逃げる
3.   何もしない（ドラゴンの炎に焼かれてゲームオーバー）





```python
#sentaku = "戦う"
#sentaku = "逃げる"
sentaku = "何もしない"
if  sentaku == "戦う":
  print("ドラゴンと戦います。何を使って戦いますか？")
elif sentaku == "逃げる":
  print("とにかく逃げましょう")
else:
  print("ドラゴンの炎に焼かれてゲームオーバー")
```

**ここで一言**：ゲームは（実は）こんな仕組みで実現されています。



遊ぶのはここまでにして、少し真面目に「プログラミングの教科書的な例題」をやってみましょう。

```python
a = 13
if a == 13:
  print("数字は13です")
```

無事に"数字は13です"と出力されました。でも、変数aが14だった場合はどうなるでしょうか？実は何も起こりません（気になる人は試してみてください）。a = 14だった場合にも何か出力したい場合、以下のようにelseを使います。

```python
a = 14
if  a == 13:
  print("数字は13です")
else:
  print("数字は13ではありません")

print("皆さん、上の行で「数字は13ではありません」と出力されましたか？")
```

次に、変数がaが（１）13のときには「数字は13です」、（２）14のときには「数字は14です」、（３）それ以外のときには「数字は13でも、14でもありません」と出力されるプログラムを作ってみましょう。
<br>

> プログラムが得意な人は「簡単だよ」と思うかもしれませんし、プログラムをやったことがない人は「え？プログラム？」とびっくりするかもしれません。でも、大丈夫です。else ifという構文を使えばとても簡単です。ちょっとやってみましょう。



```python
a = 13
if  a == 13:
  print("数字は13です")
elif a == 14:
  print("数字は14です")
else:
  print("数字は13でも、14でもありません")
```

**ある貧乏な勇者の選択**

漫画・アニメ[「ダンジョン飯」](https://delicious-in-dungeon.com/)でわかるように「勇者（と勇者のパーティー）もご飯を食べなければ死んでしまいます。ご飯（食料）を買うにはお金が必要であるため、ある貧乏な勇者が「5日分しか食料」が用意できなかったとしよう。

その場合、ダンジョンが探索できるのは「5日だけ」になります（それ以上いると「ダンジョン飯」のようにモンスターを食べることに！）

1から5までの数字を作りましょう。すでに学んだリストを使います。

forの書き方は「for 変数 in 範囲 :」です。forとinと:（コロン）が重要です。

たとえば、「for i in [1,2,3] :」と書くと、変数iに1, 2, 3を順番に入れて繰り返します。

> 注：数字が入力される変数iはstr()という関数で数字から文字に変換する必要があります。

え？よくわからない？実際にやってみましょう！

```python
suuji = [1, 2, 3, 4, 5]
print(suuji)
```

```python
for i in suuji:
  print("今日は、ダンジョン探索" + str(i) + "日目です")
```

**for文を理解しよう！**
<br>
次に0から20までの数字（整数）を順番に出力することを考えましょう。
<br>実は「単に数字を表示するだけ」ならとても簡単で、numpyのarrangeというメソッドを使います（組み込み関数のrange()を使っても構いません）。

```python
import numpy as np
np.arange(20)
```

np.arange(20)と入力すると0から19までの「合計20個の整数」が表示されることに気をつけてください（1から20ではありません！）。

もし、1から20までを表示したい場合は、最初に始める数字（この場合1）、次に終わりの数字の次の数字（この場合21）を入力します（20ではなく、21を入力するのは少し奇妙だと思うかもしれません。実は私[矢野]も少しそう思いますが、仕様で決まっているので、従うしかありません）。

```python
np.arange(1,21)
```

次に「数字は『』です」と表示させる事を考えましょう。この場合、for文（もしくはfor loop）を使います。
<br>下のfor (i in 1:20)の部分は「変数iに1から20まで順番に数字（整数）を入力する」ことを意味します。変数iはprint()の中にも出てきます。この変数iには1から20までの数字がそれぞれ入ります。やってみましょう。

> 数字が入力される変数iはstr()という関数で数字から文字に変換する必要があります。



```python
for i in np.arange(1, 21):
  print("数字は"+ str(i) + "です")
```

**応用問題をやってみよう！**
<br>
ここまで学んだifとforを組み合わせて、「1から20までの数字を順番に発生させて、（１）13のときには『数字は13です』、（２）14のときには『数字は14です』、（３）それ以外のときには『数字は13でも、14でもありません』と出力されるプログラム」を作ってみましょう

```python
for i in np.arange(1,21):
  if i == 13:
    print("数字は13です")
  elif i== 14:
    print("数字は14です")
  else:
    print("数字は13でも、14でもありません")
```

**whileにチャレンジしよう！**

「ifとforだけでもうお腹いっぱい。これ以上は無理」と思う人がいるかもしれません。実際、はじめてプログラミングをする人にはifとforだけでもかなり複雑ですね。

なので、ここから先は「頑張れる気力が残っていたら」チャレンジしてみてください（後で元気が出たときにwhileだけ勉強するというのもいいかもしれません）。



```python
i = 1
while i <= 10:
  print("数字は"+ str(i) + "です")
  i = i + 1
```
