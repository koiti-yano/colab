**足し算する関数を書こう！**

単に変数aと変数bの足し算を行い、結果（a+b）を返すというとても簡単な関数を考えましょう。「はじめて（プログラムで）関数を書く」という人もいるかもしれません。どのようなことが必要か最初にまとめてみましょう。

関数に必要なことをまとめると以下となります。
1. 変数aと変数bを関数に入力する
2. 関数の中でa+b を実行する  
3. a+bの結果を出力する（kekka=a+bとします）

実際に「足し算する関数」を書いたら以下のようになります。

````python
def tashizan(a, b):
    kekka = a + b
    return kekka

a = 1
b = 2

tashizan(a,b)
````

「aが1, bが2」ですから、a+b=3ですね。tashizan()という関数にa, bを入れる、つまりtashizan(a, b)とすると「3」という結果が帰ってきます。どうやら正しくtashizan()という関数が動いているようです。

関数の書き方に少し説明が必要でしょう。先ほど入力したtashizanという関数を再掲します。

```
def tashizan(a, b):
    kekka = a + b
    return kekka
```
関数について以下のようなルールがあります
1. 「def tashizan(a,b):」の部分はtashizanという名前の関数を作りますと宣言しています（コロンをつけるのを忘れないでください）
2. 「(a, b)」の部分は、入力する変数がaとbであることを示しています。
3. def tashizan(a,b):の下に処理の内容を書きます
4. return kekkaは出力する変数がkekkaであることを意味しています


**デフォルト引数値を設定する**

デフォルト引数値（default parameter values）を設定することができます（下の例だとb=3となっています）。デフォルト引数値を設定した場合、何も入力しなければ、その引数ではデフォルト引数値が自動的に使われます。（デフォルト値が設定されていない）引数にはちゃんと数値を入力する必要があります。

Pythonのデフォルト引数値には以下のようなルールがあります。

1.   デフォルト引数値の指定は「parameter = デフォルトの値」と書く
2.   デフォルト引数値が指定された引数のあとの引数はデフォルト引数値の指定が必要である

正確な内容は以下のURLがを参照してください。<br>https://docs.python.org/ja/3/reference/compound_stmts.html#function-definitions




````python
def tashizan2(a, b=3):
    kekka = a + b
    return kekka
````

````python
tashizan2(10)
````

````python
tashizan2(10, 5)
````

**Pythonの「辞書」を学ぶ**

　Pythonには「辞書」がある、と聞くと誰でもびっくりするでしょう。実は「辞書」という名前の仕組みがあります。「辞書」はキー (key) と値 (value) がセットになったデータの表し方です。

「なんだか、面倒なのが始まったな」と思う人がいるかも知れませんが、実はこのようなデータは皆さんの生活でよく見るものです。


**マクドナルドの単品価格**
　たとえば、マクドナルドに行ってみましょう。「ビッグマックは1個480円」、「マックシェイク（バニラ）は1つ150円」などとなっています（マクドナルドではセットしか頼まない？でも、今日は我慢して単品で注文してください）。表にすると以下のようになります。

|マックシェイク（バニラ）|150|

ビックマックがkey（キー、鍵です）、480円がvalue（値、価格）です。この表をPythonの辞書で表してみましょう。keyとvalueの間にコロンをいれ、それぞれの組み合わせをカンマで区切ります。

具体的には以下のようになります。


````python
mcdonalds = {'ビッグマック':480, 'チキンフィレオ':410, 'コカ・コーラ':120, 'スプライト':120}
````

**お客様が注文した商品の合計金額を計算しよう**

最後に「お客様が注文した商品の合計金額を計算する」関数を作ってみましょう。最初にお客様がビッグマックもしくはフィレオフィッシュを注文したとき、注文個数を掛け算して、合計金額を計算する関数を作ってみましょう。

````python
# マクドナルドの注文を処理する関数（バージョン1）
def goukei_version1(food, food_kosuu):
  #フードのメニューを辞書で作ります
  food_menu = {'ビッグマック':480, 'チキンフィレオ':410}

  # 次に食べ物の価格と個数(food_kosuu)を使って食べ物の金額を計算します
  food_price = food_menu[food] * food_kosuu

  # 合計金額がfood_priceなので、それを返します
  return f'合計金額は{food_price}円です。'
````

````python
print(goukei_version1('ビッグマック', 2))
# ビッグマック480✕2=960円になります
````

**より複雑な関数を作ってみる（頑張れる人だけ）**

ビッグマックだけだと喉が渇くので、ドリンクもほしいですね。ドリンクも入れて合計金額を計算してくれる関数はできないでしょうか？

ちょっと複雑になりますが、フード（食べ物、ビッグマックなど）とドリンク（コーラなど）の辞書を別々に作ったらできるかもしれません。


````python
# マクドナルドの注文を処理する関数（バージョン２）
def goukei_version2(food, food_kosuu, drink, drink_kosuu):
  #最初にフードのメニューを辞書で作ります
  food_menu = {'ビックマック':480, 'チキンフィレオ':410}

  # まず食べ物の価格と個数(food_kosuu)を使って食べ物の金額を計算します
  food_price = food_menu[food] * food_kosuu

  #次にドリンクのメニューを辞書で作ります
  drink_menu = {'コカ・コーラ':120, 'スプライト':120}

  # 次に飲み物の価格と個数(drink_kosuu)を使って飲み物の金額を計算します
  drink_price = drink_menu[drink] * drink_kosuu

  # 最後に食べ物の金額と飲み物の金額を足すと合計金額です
  total = food_price + drink_price
  return f'合計金額は{total}円です。'
````

````python
# 関数を呼び出して注文を試す
print(goukei_version2('ビックマック', 1, 'コカ・コーラ', 1))
# ビッグマック480+コカ・コーラ120=600円になります
````

````python
# 関数を呼び出して注文を試す
print(goukei_version2('ビックマック', 1, 'コカ・コーラ', 2))
# ビッグマック480+コカ・コーラ(2つ）240=720円になります
````

飲食店のアルバイトで、端末にお客様が注文した商品とその個数を入力したことがある人がいるかもしれません。実はこのような仕組みで金額が計算されているのです。
