**Plotnineを使うときに気をつけること**
*   全体に小カッコ()をつける（Pythonの仕様から必要）
*   小カッコ()以外はほとんどがggplot2と同じ（Rのggplotについて書かれた記事が役に立つ）



```python
(ggplot(mtcars) + aes(x='wt', y='mpg') 
+ geom_point() + ggtitle("Scatter plot"))
```

```python
ice_df = pd.read_csv("https://www.dropbox.com/s/2ky13qi387j8b05/kion_iceCream.csv?raw=1")
ice_df.head()
```

```python
ice_df = ice_df.rename(columns={'年月': 'date', '気温': 'temperature', 'アイスクリーム支出金額': 'ice_purchase'})
ice_df = ice_df.replace('年', '/', regex=True)
ice_df = ice_df.replace('月', '', regex=True)
ice_df.head()
```

```python
( ggplot(ice_df) )
```

```python
( ggplot(ice_df) + aes(x='date', y='temperature', group=1) )
```

```python
( ggplot(ice_df) + aes(x='date', y='temperature', group=1) + geom_line()  + ggtitle("Line plot"))
```

```python
(ggplot(ice_df) + aes(x='temperature', y='ice_purchase') + geom_point()  + ggtitle("Scatter plot") ) 
```
