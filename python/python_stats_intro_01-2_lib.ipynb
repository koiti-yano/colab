**平方根を計算しよう！**

平方根を計算することを考えましょう。たとえば、4の平方根は2です。つまり、
$$\sqrt{4}=2$$
ですね。たとえば、Excelの場合=sqrt(4)と入力すれば、平方根を計算できます（平方根を英語でsquare rootといい、sqrt()という関数名は英語に由来します）。

Pythonでも試してみましょう。


```python
sqrt(4)
```

おや？NameErrorという文字が出てきました。どうやらうまく計算できなかったようです。どうしたらいいでしょうか？

実はPythonで平方根を計算するためにはライブラリというものを読み込む（importする）必要があります。

sqrt()という関数はNumPyという数値計算を効率的に行うライブラリに含まれています。

Python初心者は「ライブラリ？import?なにそれ？」と思う人が多いでしょうが、理屈は後で考えるとして、まずは試してみましょう。

```python
import numpy # importは「取り込む」という意味です。
numpy.sqrt(4) # .ピリオドを使ってmath内にあるsqrt()を呼び出します
```

ところでメソッドsqrt()を使うとき、「numpy.sqrt(4)」と入力してもいいのですが、もう少し短く書きたいという人もいるかもしれません。その場合、asというコマンドを使って短くすることができます。numplyをnpと短くして使ってみましょう。
<br>
以下のように書きます。

```python
import numpy as np # numpyをnpという名前で呼び出します。ちょっと英語っぽい
np.sqrt(4)
```

**Pythonではライブラリ（モジュール）を活用しよう！**

実はPython本体には基本機能だけが含まれており、「（基本機能以外の）便利な機能」を使おうと思ったら、ライブラリという「便利な機能の集まり」をPython本体に読み込んで使う必要があります。

この「読み込む」という作業を実行するのがimportという呪文です。

Pythonには、NumPy以外にもデータの取扱が便利になるpandas、グラフ作成に役立つmatplotlib、回帰分析のstatsmodels.api、機械学習のscikit-learn(sklearn)など多くの役立つライブラリがあります。

**ライブラリ？モジュール？パッケージ？**
Pythonを使っていると、ライブラリ・モジュール・パッケージという用語が出てきます。ライブラリの中に含まれるのがモジュール、モジュールの中に含まれるのがパッケージという関係になっているのですが、Python初心者は

1. ライブラリ・モジュール・パッケージは「便利な機能をまとめたものである」
2. それらをPythonで使うにはimportする必要がある

という2点だけ分かっていれば十分です。

```python
# 最初に重要な呪文を唱えます。
# 初心者の間は何をしているのかわからなくても構いません。
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
# 注釈：statsmodelsのメソッドを使う場合は、以下のように
# statsmodels.apiのimportが必要となります。
# import statsmodels.api as sm
print(sm.__version__)
```
