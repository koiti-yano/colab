**基本用語「文・式・条件式」**

*  文とはコンピューターへの命令です。たとえば、プログラムに「print("Hello!")」という文があったら、「Helloと表示（正確には出力）せよ」とコンピューターに命令しています。プログラムは文を複数並べたもので、基本的には上から順番に実行されます
*  式とはコンピューターへ「計算せよ」と命令する文です。たとえば、プログラムに「1+1」とあったら、「1+1を計算せよ」とコンピューターに命令しています
*  条件式とは、「真（正しい）」もしくは「偽（正しくない）」が返ってくるような式です。たとえば、「1<2」と書いた場合、「1より2のほうが大きいのは真」ですから「真」が帰ってきます。

文・式・条件式は言葉で説明すると難しいと感じるかもしれませんが、実際のプログラムを見るとそれほど難しくはありません。

**TRUEとFALSEを覚えよう！**
<br>
最初は「何に役立つのかわからない！」と思うかもしれませんが、<br>Rでは「真をTRUE」、「偽をFALSE」と表します。（「真と偽なんて哲学の用語みたいで難しい」と思う人がいるかもしれません。真＝正しい、偽＝間違っているくらいに思ってください。）

たとえば、「x = 11, y = 11, z = 22」と代入したとき、

- 「x = y」は真（xには11、yにも11が入っているので、x=yは正しい。つまり真）
- 「x = z」は偽（xには11、zには22が入っているので、x=zは間違い、つまり偽）


> 場合によってはTRUEを単にT、FALSEを単にFと略すことができる場合もあります。

ただし、「x = y（数学での表記）」のように2つの値が等しいことを表すとき、**Rでは==を使います（←ここ重要）。**

```
x <- 11
y <- 11
z <- 22
```

```
x == y
```

変数xと変数yはどちらにも11が入っていて、等しい（==）ため、TRUE（真）と表示されます。それでは変数xと変数zが等しいかを確認しましょう。

```
x == z
```

変数xには11、変数zには22が入っていて、等しくないため、FALSE（偽）と表示されます。


> ==, <, > <=, >= については第6回のfor, ifのときに解説します。

**関数（組み込み関数）を使う**
<br>
次に関数という機能を使ってみましょう。たとえば、平均（平均値）を計算することを考えましょう。Rではmean()という関数を使います（半角文字のmeanに括弧()を使います）。mean()の括弧に入る変数を引数といいます。

「関数？数学みたいだ」と思う人もいるかもしれません。実際にプログラミングの関数は数学の関数に似ています。

*   **数学の関数**：$y=f(x)$。ただし、$x$が入力変数（独立変数）で、$y$が出力変数（従属変数）。意味としては関数$f()$に$x$を入力すると$y$が出力される。
*  **プログラミングの関数**：関数mean()の()内にベクトル等の引数を入力すると、計算結果が出力される。

mean()は最初から組み込まれているので組み込み関数（built-in function）と呼ばれます。組み込み関数は多いので、最初から全部覚える必要はありません。

```
#最初に元となるデータのベクトルを作ります。
motodata <- c(1, 2, 3, 4, 5)
# このベクトルmotodateが引数（入力）です

mean(motodata)
# mean()の結果を変数に入れることもできます。
heikinvec <- mean(motodata)
heikinvec
```

上のコードを見て、「あれ？#ってなんのマーク？」と思った人がいるかもしれません。#は「ここから左はコメントである」ことを意味します。プログラムにメモなどを書きたいときに使います。

それ以外にも平方根を計算するsqrt()という関数などもあります。4の平方根は2ですね。実際にやってみましょう。

```
sqrt(4)
```

**欠損値を知ろう**
<br>データの一部が「（記録ミス・紛失等の理由で）値がない」場合、欠損値（missing values）といいます。たとえば、一郎・花子・二郎の国語のテストをしたとき、二郎のテストの点数がわからなくなった場合、「二郎の点数が欠損値」です。
<br>Rでは欠損値をNAと表します。（なお、NAは「Not Available=利用できない/入手できない」の頭文字に由来すると言われています。）

> 「え？欠損値って重要なの？それをNAで表すことって大切なことなの？」と思う人もいるかも知れませんが、現実の世界では「データを取り損ねる（もしくはデータは取ったけどなくなってしまった）」ということはよくあり、「欠損値をどう表すか」を知っておくことは大切です。

以下の例を見て使い方を知っておいてください。

```
kokugo_tensu <- c(85, 92, NA) # 一郎が85点、花子が92点、二郎が欠損値
# この平均値は？
mean(kokugo_tensu)
```

二郎の点数がNA（欠損値）なので、平均もNAになってしまいました。こういうときのやり方として「NA（欠損値）を取り除く」という方法が考えられます。na.omit()という関数があります。

> ただし、「NA（欠損値）を取り除く」という方法がいつでも望ましいわけではありません。「NAを取り除くというやり方ではうまくいかない」場合もあります。興味があったら調べてみてください。



```
# 第1の方法
na.omit(kokugo_tensu)
mean(na.omit(kokugo_tensu))
# 第2の方法
mean(kokugo_tensu, na.rm=TRUE)
```
