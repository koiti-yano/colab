**足し算する関数を書こう！**

単に変数aと変数bの足し算を行い、結果（a+b）を返すというとても簡単な関数を考えましょう。「はじめて（プログラムで）関数を書く」という人もいるかもしれません。どのようなことが必要か最初にまとめてみましょう。

関数に必要なことをまとめると以下となります。
1. 変数aと変数bを関数に入力する
2. 関数の中でa+b を実行する  
3. a+bの結果を出力する（kekka=a+bとします）

実際に「足し算する関数」を書いたら以下のようになります。

````python
def tashizan(a, b):
    kekka = a + b
    return kekka

a = 1
b = 2

tashizan(a,b)
````

「aが1, bが2」ですから、a+b=3ですね。tashizan()という関数にa, bを入れる、つまりtashizan(a, b)とすると「3」という結果が帰ってきます。どうやら正しくtashizan()という関数が動いているようです。

関数の書き方に少し説明が必要でしょう。先ほど入力したtashizanという関数を再掲します。

```
def tashizan(a, b):
    kekka = a + b
    return kekka
```
関数について以下のようなルールがあります
1. 「def tashizan(a,b):」の部分はtashizanという名前の関数を作りますと宣言しています（コロンをつけるのを忘れないでください）
2. 「(a, b)」の部分は、入力する変数がaとbであることを示しています。
3. def tashizan(a,b):の下に処理の内容を書きます
4. return kekkaは出力する変数がkekkaであることを意味しています


**デフォルト引数値を設定する**

デフォルト引数値（default parameter values）を設定することができます（下の例だとb=3となっています）。デフォルト引数値を設定した場合、何も入力しなければ、その引数ではデフォルト引数値が自動的に使われます。（デフォルト値が設定されていない）引数にはちゃんと数値を入力する必要があります。

Pythonのデフォルト引数値には以下のようなルールがあります。

1.   デフォルト引数値の指定は「parameter = デフォルトの値」と書く
2.   デフォルト引数値が指定された引数のあとの引数はデフォルト引数値の指定が必要である

正確な内容は以下のURLがを参照してください。<br>https://docs.python.org/ja/3/reference/compound_stmts.html#function-definitions




````python
def tashizan2(a, b=3):
    kekka = a + b
    return kekka
````

````python
tashizan2(10)
````

````python
tashizan2(10, 5)
````
