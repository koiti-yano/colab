**「等しい」を表すには？**
たとえば、変数aに5を代入して、その後、「あれ？aは5で間違いなかったかな？」と疑問に思ったとしましょう。どのように確認したらいいでしょうか？Rではある値について書くにする記号がいくつかあります。
<br>
==は「（左右の）両辺が等しい」を意味します。ちょっと使ってみましょう。
など

```
a <- 5
a == 5
```

TRUEと表示されたのは「（変数aに5が代入されているので）a==5は正しい（TRUE。TRUEは正しい、もしくは真であるなどといいます）」ことを意味します。それ以外にも、<は「（右辺と比べて）左辺が小」 >は「（右辺と比べて）左辺が大」です。さらに<=「（右辺と比べて）左辺は小もしくは等しい」、>=「（右辺と比べて）左辺は大もしくは等しい」を意味します。<と>は比較的わかりやすいので、ちょっと分かりにくい「<=」を使ってみましょう。


> **やってみよう！** ==, <, > <=, >=それぞれを自分でやってましょう！



```
a <- 5 
b <- 6
c <- 7
a <= 6
b <= 6
c <= 6
```

実行してみたところ、「c <= 6」のときにFALSEが出てきます（FALSEは間違い・偽などといいます）。

**if文を理解しよう**
<br>
==, <, > <=, >=の使い方が分かったら、次はif文を理解しましょう。

```
a <- 13
if ( a == 13){
  cat("数字は13です")
}
```

無事に"数字は13です"と出力されました。でも、a <- 14だった場合はどうなるでしょうか？実は何も起こりません（気になる人は試してみてください）。a <- 14だった場合にも何か出力したい場合、以下のようにelseを使います。

```
a <- 14
if ( a == 13){
  cat("数字は13です")
} else {
  cat("数字は13ではありません")
}
cat("皆さん、上の行で「数字は13ではありません」と出力されましたか？")
```

次に、変数がaが（１）13のときには「数字は13です」、（２）14のときには「数字は14です」、（３）それ以外のときには「数字は13でも、14でもありません」と出力されるプログラムを作ってみましょう。
<br>

> プログラムが得意な人は「簡単だよ」と思うかもしれませんし、プログラムをやったことがない人は「え？プログラム？」とびっくりするかもしれません。でも、大丈夫です。else ifという構文を使えばとても簡単です。ちょっとやってみましょう。



```
a <- 13
if ( a == 13){
  cat("数字は13です")
} else if (a == 14){
  cat("数字は14です")
} else {
  cat("数字は13でも、14でもありません")
}
```

**遊んでみよう！**
<br>
上のプログラムの「a <- 13」の数字（半角英数）を書き換えて遊んでみてください。

**for文を理解しよう！**
<br>
次に1から20までの数字（整数）を順番に出力することを考えましょう。
<br>実は「単に数字を表示するだけ」ならとても簡単で、コロン:を使います（コロン:は半角英数字です）。

```
1:20
```

次に「数字は『』です」と表示させる事を考えましょう。この場合、for文（もしくはfor loop）を使います。
<br>下のfor (i in 1:20)の部分は「変数iに1から20まで順番に数字（整数）を入力する」ことを意味します。変数iはcat()の中にも出てきます。この変数iには1から20までの数字がそれぞれ入ります。やってみましょう。

```
for (i in 1:20){
  cat("数字は", i, "です\n")
}
```

**応用問題をやってみよう！**
<br>
ここまで学んだifとforを組み合わせて、「1から20までの数字を順番に発生させて、（１）13のときには『数字は13です』、（２）14のときには『数字は14です』、（３）それ以外のときには『数字は13でも、14でもありません』と出力されるプログラム」を作ってみましょう

```
for (i in 1:20){
  if ( i == 13){
    cat("数字は13です\n")
  } else if (i== 14){
    cat("数字は14です\n")
  } else {
    cat("数字は13でも、14でもありません\n")
  }
}
```
