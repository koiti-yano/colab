```python
# 最初にちょっと重要な呪文を唱えます。
# 最初は何をしているのかわからなくても構いません。
import statistics as st
import pandas as pd
```

計算のため、データ（CSVファイル）をインターネットから読み込みます。ここではread_csvというメソッドを使ってインターネット上にあるデータをPythonに読み込み、ice_dataという名前で保存します。head()というメソッドはデータを上から5行表示するメソッドです。

```python
ice_data = pd.read_csv("https://www.dropbox.com/s/2ky13qi387j8b05/kion_iceCream.csv?raw=1",encoding="utf_8")
ice_data.head()
```

次にデータフレームice_dataから気温とアイスクリーム支出金額を取り出してみましょう。たとえば、気温を取り出すにはice_data['気温']のように一重引用符を使います。
<br>
平均にはメソッドmean(), 分散にはメソッドvariance(), 標準偏差にはメソッドstdev()を使います。

```python
kion = ice_data['気温']
heikin = st.mean(kion)
bunsan = st.variance(kion)
hyojun = st.stdev(kion)

print(heikin)
print(bunsan)
print(hyojun)
```
