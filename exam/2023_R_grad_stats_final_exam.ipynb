```r
#このセルにある内容は変更しないようにしてください
require(tidyverse)
ssdse = read.csv("https://raw.githubusercontent.com/koiti-yano/colab/main/data/SSDSE-E-2023_modified.csv", encoding="utf_8")
ssdse %>% head()
```

```r
ssdse %>% names()
```

**ここからが期末レポートの課題提出に必要な部分です。**

以下のとおり、期末レポートを作成・提出してください。

1. ssdseから被説明変数を選択する
2. ssdseから説明変数を選択する（単回帰・重回帰どちらでも良い）
3. 作成したColabノートブックをメールに添付して提出する

なお、被説明変数と説明変数は必要に応じて変換して構いません。たとえば、対数変換、一人当たりに変換等。

```r
# 回帰モデル(1)を作成
# 以下の例は「やってはダメな回帰の例」です（総人口は十五歳未満人口を含んでいるため）
lm_model_1 <- lm(総人口 ~ 十五歳未満人口, data=ssdse)
# 最小二乗法で回帰モデルを推定
lm_model_1 %>% summary()
```

```r
# 回帰モデル(2)を作成
# 以下の例は「列名が間違っている例」です（列名を直すと推定できます。ただし、外国人人口は総人口に含まれます）
lm_model_2 <- lm('総人口 ~ 外国人', data=ssdse)
# 最小二乗法で回帰モデルを推定
lm_model_2 %>% summary()

```
