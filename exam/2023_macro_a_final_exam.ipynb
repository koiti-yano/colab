```python
#このセルにある内容は変更しないようにしてください
import pandas as pd
#import statsmodels.api as sm
import statsmodels.formula.api as smf

ssdse = pd.read_csv("https://raw.githubusercontent.com/koiti-yano/colab/main/data/SSDSE-D-2023_modified.csv",
                    encoding="cp932", skiprows=1)
ssdse.head()
```

```python
#このセルにある内容は変更しないようにしてください
ssdse.columns
```

ここからが期末レポートの課題提出に必要な部分です。

授業中の注意事項を着実に守って、期末レポートを作成・提出してください。

作業する場合の注意
1.   説明変数を選ぶとき一重引用符'を間違ってコピーしないこと
2.   smf.olsに説明変数をコピーするとき間違って、二重引用符を削除しないこと


```python
# 回帰モデル(1)を作成
# 睡眠を被説明変数としたとき、それをよく説明する説明変数を探索せよ
# よく説明する説明変数を見つけたら、なぜその説明変数が有効かを
# 考察してレポートに記述すること（20文字以上）
model_1  = smf.ols(" 睡眠 ~ その他 ", data=ssdse)
# 最小二乗法で回帰モデルを推定
result_1 = model_1.fit()

print(result_1.params)
print(result_1.rsquared_adj)
#result_1.summary()
```

```python
# 回帰モデル(2)を作成
# 学業を被説明変数としたとき、それをよく説明する説明変数を探索せよ
# よく説明する説明変数を見つけたら、なぜその説明変数が有効かを
# 考察してレポートに記述すること（20文字以上）
model_2  = smf.ols(" 学業 ~ 食事 ", data=ssdse)
# 最小二乗法で回帰モデルを推定
result_2 = model_2.fit()

print(result_2.params)
print(result_2.rsquared_adj)
```

```python
# 回帰モデル(3)を作成
# 受診療養を被説明変数としたとき、それをよく説明する説明変数を探索せよ
# よく説明する説明変数を見つけたら、なぜその説明変数が有効かを
# 考察してレポートに記述すること（20文字以上）
model_3  = smf.ols(" 受診療養 ~ その他 ", data=ssdse)
# 最小二乗法で回帰モデルを推定
result_3 = model_3.fit()

print(result_3.params)
print(result_3.rsquared_adj)
```
