```python
#このセルにある内容は変更しないようにしてください
import pandas as pd
#import statsmodels.api as sm
import statsmodels.formula.api as smf

ssdse = pd.read_csv("https://raw.githubusercontent.com/koiti-yano/colab/main/data/SSDSE-E-2023_modified.csv", encoding="utf_8")
ssdse.head()
```

```python
#このセルにある内容は変更しないようにしてください
ssdse.columns
```

**ここからが期末レポートの課題提出に必要な部分です。**

授業中の注意事項を着実に守って、期末レポートを作成・提出してください。

```python
# 回帰モデル(1)を作成
# 以下の例は「やってはダメな回帰の例」です（授業資料を参照してください）
sm_model_1  = smf.ols('総人口 ~ 十五歳未満人口', data=ssdse)
# 最小二乗法で回帰モデルを推定
result_1 = sm_model_1.fit()

print(result_1.params)
result_1.summary()
```

```python
result_1.rsquared_adj
```

```python
# 回帰モデル(2)を作成
# 以下の例は「列名が間違っている例」です（授業資料を参照してください）
sm_model_2  = smf.ols('総人口 ~ 外国人', data=ssdse)
# 最小二乗法で回帰モデルを推定
result_2 = sm_model_2.fit()

print(result_2.params)
```

```python
result_2.rsquared_adj
```
