**TrueとFalseを覚えよう！**
<br>
最初は「何に役立つのかわからない！」と思うかもしれませんが、<br>Pythonでは「真をTrue」、「偽をFalse」と表します。（「真と偽なんて哲学の用語みたいで難しい」と思う人がいるかもしれません。真＝正しい、偽＝間違っているくらいに思ってください。）

たとえば、「x = 11, y = 11, z = 22」と代入したとき、

- 「x = y」は真（xには11、yにも11が入っているので、x=yは正しい。つまり真）
- 「x = z」は偽（xには11、zには22が入っているので、x=zは間違い、つまり偽）

ただし、「x = y（数学での表記）」のように2つの値が等しいことを表すとき、**Pythonでは==を使います（←ここ重要）。**

```python
x = 11 # 変数xに11を代入します
y = 11 # 変数yに11を代入します
z = 22 # 変数zに22を代入します
```

```python
x == y # 変数xと変数yが等しいかどうか確認します。
```

変数xと変数yはどちらにも11が入っていて、等しい（==）ため、True（真）と表示されます。それでは変数xと変数zが等しいかを確認しましょう。

```python
x == z
```

変数xには11、変数zには22が入っていて、等しくないため、False（偽）と表示されます。


> ==, <, > <=, >= については第6回のfor, ifのときに解説します。



**リストを作ろう**<br>次にリストを作りましょう。リストは数（スカラー）等の組み合わせから作ることができます。
<br>
作り方は大括弧[]とカンマ,で数（スカラー）の組み合わせを作るだけです。リストの名前はnumlistのように複数の半角英数字を組み合わせて作ることができます。


```python
numlist = [1, -2.1, 5.6, 4]
numlist
```

一つ上で作ったリストnumlistから要素（数字）を取り出してみましょう。numlistの中で左から1番目の要素（数）は1、左からから3番目の要素は5.6です。

ここで重要なポイント：Pythonでインデックスは0から始まるので、左から1番目の要素（数）を取り出すときインデックスは0、左から3番目の要素（数）を取り出すときインデックスは2になります。また、取り出すときには[]を使います。

```python
numlist[0]
```

```python
numlist[2]
```

「Pythonでインデックスは0から始まる」というのは最初は戸惑う人がいるかももしれませんが、慣れれば難しくはありません。

**文字列を使う**
<br>
文字列（文字や文章）を変数にすることもできます。この場合、二重引用符""もしくは一重引用符''を使います。

```python
mojiretsu = "今日の天気は晴れです"
mojiretsu
```

mojiretsuという変数の中に「今日の天気は晴れです」と文字列（文章）が入っていることがわかります。文字列のリストを作ることもできます。

```python
mojilist = ["今日の天気は晴れです", "昨日の天気は雨でした", "明日の天気は予報では雪です"]
mojilist
```

文字列のリストの2番目の要素は"昨日の天気は雨でした"です。それを取り出してみましょう。この場合も[]を使います。

```python
mojilist[1]
```

**モジュールのインポート**
<br>
Pythonでは多くの便利な機能がモジュールという単位でまとめられています。モジュールはimportというコマンドでPythonに読み込んでから使います。
<br>
たとえば数学に関する便利な機能の多くはmathというモジュールに含まれています。たとえば平方根を計算するsqrt()という関数はmathに含まれています。なお、モジュールに含まれる関数はメソッドといいます。
<br>
4の平方根は2です。そうなるのか実際に試してみましょう。

```python
import math
math.sqrt(4)
```

ところでメソッドsqrt()を使うとき、「math.sqrt(4)」と入力してもいいのですが、もう少し短く書きたいという人もいるかもしれません。その場合、asというコマンドを使って短くすることができます。mathをmtと短くして使ってみましょう。
<br>
以下のように書きます。

```python
import math as mt
mt.sqrt(4)
```

Pythonには、math以外にもデータの取扱が便利になるpandasや数値計算に関するモジュールnumpy、回帰分析のstatsmodels.api、機械学習のscikit-learn(sklearn)など多くの役立つモジュールがあります。たとえば、numpyを使って、0から5までの数字（整数）の列を作ってみましょう。numpyのarangeというメソッドを使います。

```python
import numpy as np
np.arange(5)
```
