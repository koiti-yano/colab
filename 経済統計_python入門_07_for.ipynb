**「等しい」を表すには？**<br>
たとえば、変数aに5を代入して、その後、「あれ？変数aに5がちゃんと入っているかな？」と疑問に思ったとしましょう。どのように確認したらいいでしょうか？
<br>

まず、変数aに5を代入してみましょう。=を使います。

```python
a = 5
```

これで変数aに数字の5が代入されました。続いて、「変数aに数字の5が代入されているか」を確認してみましょう。

この場合==（等号を2つ）を使います。==は「（左右の）両辺が等しい」を意味します。ちょっと使ってみましょう。

```python
a == 5
```

True（真）と表示されました。これは「（変数aに5が代入されているので）a==5は正しいことを意味します。

それ以外にも、<は「（右辺と比べて）左辺が小」 >は「（右辺と比べて）左辺が大」です。さらに<=「（右辺と比べて）左辺は小もしくは等しい」、>=「（右辺と比べて）左辺は大もしくは等しい」を意味します。<と>は比較的わかりやすいので、ちょっと分かりにくい「<=」を使ってみましょう。


> **やってみよう！** ==, <, > <=, >=それぞれを自分でやってましょう！



```python
a = 5 
b = 6
c = 7
a <= 6
```

```python
b <= 6
```

```python
c <= 6
```

変数cには7が入っていますから「c <= 6」のときにはFalse（偽）と出てきます。

**if文を理解しよう**
<br>
==, <, > <=, >=の使い方が分かったら、次はif文を理解しましょう。

```python
a = 13
if a == 13:
  print("数字は13です")
```

無事に"数字は13です"と出力されました。でも、a <- 14だった場合はどうなるでしょうか？実は何も起こりません（気になる人は試してみてください）。a <- 14だった場合にも何か出力したい場合、以下のようにelseを使います。

```python
a = 14
if  a == 13:
  print("数字は13です")
else:
  print("数字は13ではありません")

print("皆さん、上の行で「数字は13ではありません」と出力されましたか？")
```

次に、変数がaが（１）13のときには「数字は13です」、（２）14のときには「数字は14です」、（３）それ以外のときには「数字は13でも、14でもありません」と出力されるプログラムを作ってみましょう。
<br>

> プログラムが得意な人は「簡単だよ」と思うかもしれませんし、プログラムをやったことがない人は「え？プログラム？」とびっくりするかもしれません。でも、大丈夫です。else ifという構文を使えばとても簡単です。ちょっとやってみましょう。



```python
a = 13
if  a == 13:
  print("数字は13です")
elif a == 14:
  print("数字は14です")
else:
  print("数字は13でも、14でもありません")
```

**for文を理解しよう！**
<br>
次に0から20までの数字（整数）を順番に出力することを考えましょう。
<br>実は「単に数字を表示するだけ」ならとても簡単で、numpyのarrangeというメソッドを使います（組み込み関数のrange()を使っても構いません）。

```python
import numpy as np
np.arange(20)
```

np.arange(20)と入力すると0から19までの「合計20個の整数」が表示されることに気をつけてください（1から20ではありません！）。

もし、1から20までを表示したい場合は、最初に始める数字（この場合1）、次に終わりの数字の次の数字（この場合21）を入力します（20ではなく、21を入力するのは少し奇妙だと思うかもしれません。実は私[矢野]も少しそう思いますが、仕様で決まっているので、従うしかありません）。

```python
np.arange(1,21)
```

次に「数字は『』です」と表示させる事を考えましょう。この場合、for文（もしくはfor loop）を使います。
<br>下のfor (i in 1:20)の部分は「変数iに1から20まで順番に数字（整数）を入力する」ことを意味します。変数iはprint()の中にも出てきます。この変数iには1から20までの数字がそれぞれ入ります。やってみましょう。

> 数字が入力される変数iはstr()という関数で数字から文字に変換する必要があります。



```python
for i in np.arange(1, 21):
  print("数字は"+ str(i) + "です")
```

**応用問題をやってみよう！**
<br>
ここまで学んだifとforを組み合わせて、「1から20までの数字を順番に発生させて、（１）13のときには『数字は13です』、（２）14のときには『数字は14です』、（３）それ以外のときには『数字は13でも、14でもありません』と出力されるプログラム」を作ってみましょう

```python
for i in np.arange(1,21):
  if i == 13:
    print("数字は13です")
  elif i== 14:
    print("数字は14です")
  else:
    print("数字は13でも、14でもありません")
```
