**足し算する関数を書こう！**

単に変数aと変数bの足し算を行い、結果（a+b）を返すというとても簡単な関数を考えましょう。「はじめて（プログラムで）関数を書く」という人もいるかもしれません。どのようなことが必要か最初にまとめてみましょう。

関数に必要なことをまとめると以下となります。
1. 変数aと変数bを関数に入力する
2. 関数の中でa+b を実行する  
3. a+bの結果を出力する（kekka=a+bとします）

実際に「足し算する関数」を書いたら以下のようになります。


````
tashizan <- function(a, b){ 
  kekka <- a + b
  return(kekka)
  }
a <- 1
b <- 2
tashizan(a,b)
````

「aが1, bが2」ですから、a+b=3ですね。tashizan()という関数にa, bを入れる、つまりtashizan(a, b)とすると「3」という結果が帰ってきます。どうやら正しくtashizan()という関数が動いているようです。

関数の書き方に少し説明が必要でしょう。先ほど入力したtashizanという関数を再掲します。

```
tashizan <- function(a, b){ 
  kekka <- a + b
  return(kekka)
  }
```
関数について以下のようなルールがあります
1.   「tashizan <- function」の部分はtashizanという名前の関数（function）を作りますと宣言しています
2.   「(a, b)」の部分は、入力する変数がaとbであることを示しています。
3. 中括弧{}で囲まれた部分に処理の内容を書きます
4. return(kekka)は出力する変数がkekkaであることを意味しています




**ちょっとした書き方（１）**

以下のように関数の結果を変数に入れることができます。

````
ke <- tashizan(a,b)
````

**ちょっとした書き方（２）**
a <- 3, b <- 4のようにせず、(a=3, b=4)のようにして関数に入力することもできます。

````
tashizan(a=3,b=4)
````

**デフォルト値を設定する**

デフォルト値を設定することができます（下の例だとa=2となっています）。デフォルト値を設定した場合、入力しなければ、そのデフォルト値が自動的に使われます。そのときには（デフォルト値が設定されていない）bにはちゃんと数値を入力する必要があります。


````
tashizan2 <- function(a=2, b){ 
  kekka2 <- a + b
  return(kekka2)
  }
tashizan2(b=3)
````

````
a <- 1
tashizan2(b=3)
# aに1が入っているが、それはtashizan2の引数aに反映されず、デフォルトのa=2が使われている。
````

(a <- 1)でaに1が代入されていますが、それがtashizan2の引数aに反映されず、デフォルトのa=2が使われているため、2+3=5という結果が帰ってきました。

ここでパイプ演算子|>の出番です。

**復習：パイプ演算子|>の機能**：|>の左辺のデータを、|>の右辺の関数の第一引数に入力する


> 補足：2つのパイプ演算子：Rで使えるパイプ演算子としては%>%と|>があります。%>%はtidyverseにあるmagrittrというパッケージのパイプ演算子で広く使われてきました。しかし、R 4.1.0以降、R本体に|>というパイプ演算子が浸かるようになりました。%>%と|>の機能は非常に似ているので、どちらを使っても構いませんが、今後は|>が使われるようになってくるのではないかと思います。


````
a <- 1
a |> tashizan2(b=3)
````

今度は(a <- 1)でaに1が代入され、それがパイプ演算子|>の機能でtashizan2の引数aに代入され1+3=4という結果が帰ってきました。

補足：パイプ演算子には他にplaceholderという機能もあります。パイプ演算子については、改めて他の資料で解説します。
