#**matplotlibでグラフ作成**

matplotlibは機能が豊富でいろいろなことができますが、最初は折れ線グラフのためにメソッドplot()、散布図作成にメソッドscatter()を覚えましょう。

> 以下のサイトを参考にしました。<br>Usage Guide(Matplotlib公式サイト)
<br>
https://matplotlib.org/stable/tutorials/introductory/usage.html


```python
# 最初にちょっと重要な呪文を唱えます。
# 最初は何をしているのかわからなくても構いません。
import matplotlib.pyplot as plt
import pandas as pd
```

グラフを作るため、データ（CSVファイル）をインターネットから読み込みます。ここではread_csvというメソッドを使ってインターネット上にあるデータをPythonに読み込み、ice_dataという名前で保存します。head()というメソッドはデータを上から5行表示するメソッドです。

```python
ice_data = pd.read_csv("https://www.dropbox.com/s/2ky13qi387j8b05/kion_iceCream.csv?raw=1",encoding="utf_8")
ice_data.head()
```

次にデータフレームice_dataから気温とアイスクリーム支出金額を取り出してみましょう。たとえば、気温を取り出すにはice_data['気温']のように一重引用符を使います。
<br>
ice_data['気温']をkionという変数に代入したあと、plt.plot()を使って気温のデータを折れ線グラフにします。

```python
kion = ice_data['気温']
ice_shishutu = ice_data['アイスクリーム支出金額']
plt.plot(kion)
plt.show()
```

これだけだと横軸・縦軸のラベルがないので、少し不親切です。残念ながらColabのグラフは日本語に対応していないため日本語のラベルは付けらなのですが、英語でちょっとラベルを付けましょう。

```python
kion = ice_data['気温']
ice_shishutu = ice_data['アイスクリーム支出金額']
plt.plot(kion)
plt.title('Temparature')
plt.xlabel('Year/Month')
plt.ylabel('Ice cream')
plt.show()
```

次に気温とアイスクリーム支出金額の２つのデータを使って散布図を作ってみましょう。その場合、plt.scatter()を使います。

```python
plt.scatter(kion,ice_shishutu)
plt.xlabel('Temparature')
plt.ylabel('Ice creame')
plt.show()
```
