```r
library(jsonlite)
```

```r
zip_code <- 1500043
zip_url = paste("https://zipcloud.ibsnet.co.jp/api/search?zipcode=", zip_code, sep="")
data = read_json(zip_url)
address = data$results[[1]]
cat(address$address1, address$address2, address$address3, "\n")
```
