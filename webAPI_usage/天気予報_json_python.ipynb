#**Pythonで気象庁のWebAPIを使う**
スマホのアプリで今日の天気や明日の天気を見たことがある人もいるかと思います。スマホのアプリは（基本的には）気象情報会社から提供される情報をネットから取ってきて表示しています。

実際の仕組みとは少し違うかもしれませんが、Pythonとrequestsを使って気象庁のWeb API（気象庁はWeb APIと呼んでいないようですが）から今日・明日・明後日の天気予報を取得し、表示することを考えてみましょう。

```python
import requests as rq 
import pprint as pp
```

```python
# 東京
area_index = "130000"
forecast_url = "https://www.jma.go.jp/bosai/forecast/data/forecast/" + area_index +".json"
forecast_jma = rq.get(forecast_url).json()
type(forecast_jma)
```

```python
# pprint()を使うと出力が比較的読みやすく
# https://docs.python.org/ja/3/library/pprint.html
#(forecast_jma)
pp.pprint(forecast_jma)
```

```python
area_name = forecast_jma[0]['timeSeries'][0]['areas'][0]['area']['name']
print(area_name)

weathers = forecast_jma[0]['timeSeries'][0]['areas'][0]['weathers']

print("今日の天気：" + weathers[0])
print("明日の天気：" + weathers[1])
print("明後日の天気：" + weathers[2])
```
