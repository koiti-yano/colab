#**Pythonで郵便番号検索APIを使う**
Pythonとrequestsを使って郵便番号検索API（zipcloud）を利用する方法を解説します。

Webで会員登録などで住所を入力するとき、郵便番号を入れると住所が自動で入力されることがあります。たとえば、郵便番号150-0043を入力すると東京都・渋谷区・道玄坂と自動で入力されるようなことが考えられます（郵便番号150-0043は渋谷のスクランブル交差点がある場所です）。

この自動入力機能はWeb APIを使うと実現できます。以下では郵便番号検索API（zipcloud）を使った例を解説します。

```python
import requests as rq
import pprint as pp
```

```python
zip_code = "1500043" #渋谷のスクランブル交差点の郵便番号
zip_url = "https://zipcloud.ibsnet.co.jp/api/search?zipcode=" + zip_code
addr_data = rq.get(zip_url).json()

type(addr_data)
```

```python
# 以下の場合は比較的簡単な構造であるため、単にaddr_dataと
# してもあまり変わらないが、一応pprint()を使う
pp.pprint(addr_data)
```

```python
address = addr_data["results"][0]
print(address["address1"], address["address2"], address["address3"])
```
