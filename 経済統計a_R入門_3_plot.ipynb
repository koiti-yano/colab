**グラフを描く準備**
<br>
グラフを書く準備をします。まず、seq()という関数の使い方から始めましょう。seq()はたとえば、-6, -5.5, …, 0, …, 5.5, 6という0.5刻みの数字の組み合わせ（ベクトル）を作りたいとします。
<br>
その場合、seq(from=-0.5, to=0.5, by=0.5)と書きます。fromが始まる数字、toが終わる数字、byが刻みを表します。
<br>ちなみにfrom, to, byを省略して簡単にseq(-6, 6, 0.5)と書くこともできるのですが、最初のうちはfrom, to, byを省略せずに書くことを勧めます。



```
x <- seq(from=-6, to=6, by=0.5)
x
```


<br>
次に登場するsin()という関数は正弦sinを出力する関数です。この結果をplot()に入れることでグラフを描くことができます。

> ちなみにちょっと変わった書き方としてplot(y <- sin(x))と書くこともできます。








```
plot(sin(x))
# もしくは
y <- sin(x)
plot(y)
```

でも、これだとグラフが点（白丸）になっていますね。できれば折れ線グラフにしたいところです。その場合、以下のようにします。type="l"という部分がグラフを「線」にすることを意味しています。注：線＝lineなので"l"と覚えてください。

```
plot(sin(x), type="l")
```
