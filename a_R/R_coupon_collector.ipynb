```r
num_epoch <- 10000
max_trial <- 100

coupon_obtained <- array(NA, c(num_epoch))
#coupon_array <- array(NA, c(num_trial, num_epoch))
```

```r
for(ii in 1:num_epoch){
  for(jj in 1:max_trial){
    result <- sample(x=1:6, size=1)
    if(result==1){
#      cat("Result", result, "\n")
      coupon_obtained[ii] <- jj
#     cat("Coupon obtained: ", coupon_obtained[ii], "\n")
      break
    }
  }
}
mean(coupon_obtained)
```
