（すぐに終わりたい人は「後」を選択してください。ドラゴンやスライムと戦うと永遠に終わりません！）。

```python
# This game is originally written by https://trinket.io/
# See https://trinket.io/python/e5a03e7cbc
import time

# 準備
yes_no = ["はい", "いいえ"]
directions = ["左", "右", "前", "後"]

# 導入
print("This game is originally written by https://trinket.io/")
print("See https://trinket.io/python/e5a03e7cbc")

print("================================================")
print("アドベンチャー・ゲームへようこそ！")
print("================================================")

name = input("最初にあなたの名前を入力してください\n")
print(name + "さん、こんにちは！冒険を始めましょう！あなたはダンジョン（迷宮）の入り口にいます。")

# ゲーム開始
response = ""
while response not in yes_no:
    response = input("ダンジョンに入りますか？ \n はい/いいえ \n")
    if response == "はい":
        print("あなたはダンジョンに入りました\n")
    else:
        print("本当は「はい」と回答しないとダンジョンに入ることはできませんが、今回は特別に" + name + "さんを入れてあげましょう\n")

# ダンジョンに入ってから
f_flag = 0
response = ""
while response not in directions:

    if f_flag == 0: # ダンジョンに入ったときだけ表示
      f_flag = 1
    elif f_flag == 1:
      print("")
      print("左にドラゴンがいます。")
      print("右にはダンジョンが続いています。")
      print("前はスライムです。")
      print("後はダンジョンから出ます。\n")
      response = input("どちらに進みますか? \n 左/右/前/後 \n")
    print("===============================\n")
    if response == "左":
      print("⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⢀⣶⠀⠀⢀⣄⠀⠀⣠⣶⣾⠇⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀")
      print("⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⢀⣼⡛⣆⣰⣿⣿⣠⠞⣓⣿⣿⠶⠞⠛⣫⣿⣷⡆⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀")
      print("⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⣸⣿⡥⣿⡏⣸⡿⠛⠉⠉⠉⠉⠉⠓⢲⣿⠿⢱⣿⢤⣀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀")
      print("⠀⠀⠀⠀⠀⠀⠀⢀⡴⠶⢿⡋⠀⠟⠛⠁⣀⣀⣀⠀⠀⠀⠺⡷⠚⠉⢀⣾⣿⣶⣿⠗⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀")
      print("⠀⠀⠀⠀⠀⠀⠀⢸⣿⣷⣅⠙⠀⠀⢠⠞⢛⣿⣭⣙⠛⣦⡀⠹⣄⣀⡼⣻⣿⣯⣀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀")
      print("⠀⠀⠀⠀⠀⠀⠀⢸⣿⡏⣿⠀⠀⠀⢠⡞⠋⢹⡟⠟⢳⡈⢧⠀⠈⠙⠿⢻⠃⣷⠈⢻⣄⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀")
      print("⠀⠀⠀⠀⠀⠀⣀⣸⡿⠿⠟⠲⠶⢤⣼⠀⢹⣿⠁⠀⢨⡇⠀⠀⠀⠀⠀⠉⠀⢿⣷⡾⣿⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀")
      print("⠀⠀⠀⠀⠀⡾⠹⠆⠀⠀⠀⠀⠀⠀⠈⠳⣼⣿⣷⣤⡾⠁⠀⠀⠀⠀⠀⠀⠀⣿⣏⠻⠟⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀")
      print("⠀⠀⠀⠀⠘⣧⣀⡀⠀⠀⠐⠓⠀⠀⠀⠀⠀⠀⠀⠉⠀⠀⠀⠀⠀⠀⠀⠀⣰⠃⠘⣧⠀⠀⠀⠀⢠⣄⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀")
      print("⠀⠀⠀⠀⠀⠈⢯⡉⠙⢦⣀⠀⠀⠀⠀⠀⢀⣰⠏⠀⠀⠀⠀⠀⠀⠀⠀⣶⣯⣤⢄⡿⠀⢀⣀⣠⣾⣏⠳⣄⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀")
      print("⠀⠀⠀⠀⠀⠀⠈⠹⢦⣀⣉⠒⠶⠶⠶⢶⣊⣡⣄⣀⣀⣀⣀⡤⠀⠀⠀⡿⠙⣯⣹⠷⣚⣋⣉⣡⡴⠟⢦⠈⣷⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀")
      print("⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠉⠙⠛⠛⠛⠛⠉⠀⠀⢹⣯⠉⠁⢠⣄⡀⠀⢤⣤⣬⣿⣟⠉⠉⠁⢠⠀⠀⠀⢳⡸⡆⠀⠀⠀⠀⠀⠀⠀⠀⠀")
      print("⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⢀⡾⠋⠀⠀⠀⢺⡇⠙⢷⣽⣧⣠⣿⠿⢿⡉⠻⣾⣤⢤⣄⠀⣧⡇⠀⠀⠀⠀⠀⠀⠀⠀⠀")
      print("どうしますか？戦う/逃げる\n")
      whattodo = input("\n")
      if whattodo =="戦う":
        print("戦って勝ちました\n")
      else:
        print("逃げましょう\n")

      response = ""
      #quit()
    elif response == "右":
      print(name + "さんはダンジョンを進みました。\n")
      response = ""
    elif response == "前":
      print("　　　　 人　 　　　　 人　 　　　　 人　 　　　　 人  ")
      print("　 　 (　ﾟーﾟ) 　 　 (　ﾟーﾟ) 　 　 (　ﾟーﾟ) 　 　 (　ﾟーﾟ) ")
      print("スライムと戦って勝ちました。\n")
      response = ""
    elif response == "後":
      print("ダンジョンから出ます。" + name + "さん、さようなら！\n")
      quit()
    else:
      print("回答は左/右/前/後だけにしてください。\n")
```
