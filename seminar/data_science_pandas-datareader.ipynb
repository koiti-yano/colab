**FREDとは？**
<br>
FRED（Federal Reserve Economic Data）とはアメリカのセントルイス連邦準備銀行が提供している経済統計データベースです。このサイトではアメリカを含め世界各国の経済統計（経済データ）約80万種類が提供されています。提供しているデータはすべてオープンデータです。

**公式サイト**
<br>
https://fred.stlouisfed.org/

```python
# 最初に重要な呪文を唱えます。
# 初心者の間は何をしているのかわからなくても構いません。
import pandas_datareader as pdr
import datetime as dt
```

日経平均（日経平均株価指数）をFRED（インターネット）から読み込みます。ここではDataReader()というメソッドを使ってFRED上にあるデータをPythonに読み込み、nikkei225_1という名前で保存します。

head()というメソッドはデータを上から5行表示するメソッドです。

```python
nikkei225_1 = pdr.DataReader(name="NIKKEI225", data_source="fred")
nikkei225_1.head()
```

```python
nikkei225_1.plot()
```

データをダウンロードするときの開始時期（start）と終了時期（end）を指定します。

```python
date_start = dt.date(2010,1,1)
date_end = dt.date(2021,3,31)
nikkei225_2 = pdr.DataReader(name="NIKKEI225", data_source="fred", start=date_start, end=date_end)
nikkei225_2.plot()
```
