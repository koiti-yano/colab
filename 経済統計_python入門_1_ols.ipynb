```python
# 最初にちょっと重要な呪文を唱えます。
# 最初は何をしているのかわからなくても構いません。
import numpy as np
import pandas as pa
import statsmodels.api as st

```

```python
df = pa.read_csv("https://www.dropbox.com/s/2ky13qi387j8b05/kion_iceCream.csv?raw=1",encoding="utf_8")
df.head()
```

```python
x = df['気温']
y = df['アイスクリーム支出金額']
x1 = st.add_constant(x)

# 回帰分析実行
st_model = st.OLS(y, x1)
result = st_model.fit()

print(result.params)


```

add_constant()がカギ

```python
print(result.summary())
```
