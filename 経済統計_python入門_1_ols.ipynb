```python
# 最初にちょっと重要な呪文を唱えます。
# 最初は何をしているのかわからなくても構いません。
import numpy as np
import statsmodels.api as st
import csv
```

ダミーデータを作って、回帰する。add_constant()がカギ

```python
X = np.random.uniform(-10, 10, size=(100, 4))
y = 5 + X@[2, -3, 4, 0] + np.random.normal(0, 1, 100)
np.savez("sample_data.npz", X=X, y=y)

# データの読み込み
npzfile = np.load("sample_data.npz")
X = npzfile["X"]
y = npzfile["y"]

# 配列Xに1だけからなる列を追加する。
X0 = st.add_constant(X)
# 回帰分析実行
st_model = st.OLS(y, X0)
results = st_model.fit()

print(results.params)

print(results.summary())
```
