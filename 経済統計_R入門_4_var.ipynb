**データの準備**
<br>
*今回もまず使用するデータを準備することから始めましょう（「read.csvには慣れてきた」という人がいるかもしれません）。

```
ice.data <- read.csv("https://www.dropbox.com/s/2ky13qi387j8b05/kion_iceCream.csv?raw=1",header=T)
 head(ice.data)
```

**data.frameからデータ（列）を取り出す**
<br>

```
head(ice.data$気温)
head(ice.data$アイスクリーム支出金額)
```

**Rで平均・分散・標準偏差の計算**
<br>
平均はmean()、分散［不偏分散］はvar()、標準偏差［不偏標準偏差］はsd()という関数を使います。

```
mean(ice.data$気温)
var(ice.data$気温)
sd(ice.data$気温)
```

**Rで中央値の計算**
<br>
中央値はmedian()という関数を使います。

```
median(ice.data$気温)
```
