```python
# 最初に重要な呪文を唱えます。
# pandasというモジュールをpdという名前で読み込みます。
import pandas as pd
#import statistics as st
```

授業で使った「アイスクリーム屋の売上と気温」のデータをインターネット（dropbox）からダウンロードします。
<br>
Pandasのメソッドであるread_csv()を使って、ice_dataという変数に代入します。head()は最初の5行を表示するメソッドです。

```python
ice_data = pd.read_csv("https://www.dropbox.com/s/teoo1n1gxrlyjbd/icecream_uriage_temperature.csv?raw=1",encoding="utf_8")
ice_data.head()
```

まず、散布図を作成します。plot.scatter()というメソッドを使います。


```python
ice_data.plot.scatter(x='気温',y='売り上げ')
```
