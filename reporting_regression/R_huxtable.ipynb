```
lm.res1 <- lm(mpg ~ wt , data = mtcars) 
summary(lm.res1)
```

```
lm.res2 <- lm(mpg ~ wt + qsec, data = mtcars)
summary(lm.res2)
```

```


lm.hux <- huxreg(lm.res1, lm.res2, number_format = 3)
lm.hux
```
