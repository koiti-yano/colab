```python
# 最初に重要な呪文を唱えます。
# pandasというモジュールをpdという名前で読み込みます。
import pandas as pd
#import statistics as st
```

授業で何度も使った「小さな町の世帯年収」のデータをインターネット（dropbox）からダウンロードします。
<br>
Pandasのメソッドであるread_csv()を使って、incomesという変数に代入します。head()は最初の5行を表示するメソッドです。

```python
incomes = pd.read_csv("https://www.dropbox.com/s/b1m49lng82mt779/incomes_in_a_small_town.csv?raw=1",encoding="utf_8")
incomes.head()
```

まず、標本標準偏差を計算します。std()というメソッドを使います。ただし、ddof=0というオプションをつけてください。

**気をつけるポイント**
*   ddof=0をつける→標本標準偏差になる
*   ddof=0をつけない→不偏標準偏差になる



```python

incomes.std(ddof=0)
```

ddof=0をつけると、講義資料と同じ結果になりましたね。

次は不偏標準偏差です。同じstd()というメソッドを使いますが、オプションは何もつけません。

```python
incomes.std()
```
