重回帰（回帰分析）のため、データ（CSVファイル）をインターネットから読み込みます。ここではread.csv()という関数を使ってインターネット上にあるデータをRに読み込みます。head()は既に解説しました。

> 下記のデータはgampinderから



```
#古い toukei.data <- read.csv("https://www.dropbox.com/s/y2uvw3xv1lp3vca/toukei_multireg_data.csv?raw=1",header=T)
toukei.data <- read.csv("https://raw.githubusercontent.com/koiti-yano/colab/main/data/toukei_multireg_data.csv",header=T)
head(toukei.data)
```

**重回帰をやってみよう**<br>
最後にtoukei.dataを使って重回帰（回帰分析）をやってみましょう。単回帰と同じくlm()を使います。

**lm()の使い方**<br>
m()の使い方は、単回帰の場合と基本的な考え方は同じです。
<br>必要なのは(1) 回帰式を指定するところ、(2) 使用するデータフレーム名を指定するところの2箇所だけです。

既に解説したとおり、関数lm()の結果も変数に入力することができますので、回帰結果をlm.resultとしてみましょう。

```
lm.result <- lm(平均寿命 ~ 一人当たりのGDP + GDPに占める医療費の割合, data=toukei.data)
summary(lm.result)
```

summary(lm.result)の読み方は授業資料で解説します。これで回帰分析（lm()）の大雑把な使い方は終わりです。
